### To find top ten investment themes 

```python
WorldBanks_json.mjtheme_namecode.head()
```

```python
def Dataframe_list_access() :                               #Function to flatten JSON data into a list
    x = []
    for item in WorldBanks_json.mjtheme_namecode :
        for dic in item :
            x.append(dic['name'])
    return x
```

```python
y = Dataframe_list_access()
y                                                           #View of project themes with missing values
```

```python
y = pd.Series(y).value_counts()
```

```python
y
```

#### The Dataframe seems to have missing data. 
#### To fill in missing data by interpolating from existing data :

```python
def Dataframe_list_to_dictionary() :                          #Function to create a master dictionary of themes 
    d = {}
    for i in WorldBanks_json.mjtheme_namecode :
        for j in i :
            if j['name'] != '' :
                d[j['code']] = j['name']
    return d
```

```python
masterDictionary = Dataframe_list_to_dictionary()
masterDictionary
```

#### The master dictionary helps match missing theme name using available theme code. 
#### To fill the missing values :

```python
WB = pd.Series(WorldBanks_json.mjtheme_namecode)
WB.head()                                                   #A copy of mjtheme_namecode column
```

```python
#Uncomment to see the initial value of the 'name' for a missing value
#print(WB[0][1]['name']
```

```python
for list_of_dicts in WB :
    for dic in list_of_dicts:
        if dic['name'] == '' :
            code_ref = dic['code']
            dic['name'] = masterDictionary[code_ref]       #Matching from masterDictionary to fill in missing values
```

```python
#Uncomment to see change after the function has been run
#WB[0][1]['name']
```

```python
WorldBanks_json.mjtheme_namecode = WB                      #Replace raw column with WB
```

```python
y = Dataframe_list_access()
y                                                          #View of project themes with filled in values
```

```python
y = pd.Series(y).value_counts()
```
